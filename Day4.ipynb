```python
# Take an integer n and print its multiplication table up to 10.
n=int(input("Enter a number:"))
for i in range (1,11):
  a=n*i
  print(a)
```

```python
# Take an integer n and find the sum of its digits.
n=int(input("Enter a number:"))
sum=0
for i in range(1,n+1):
  sum+=i
print(sum)
```

```python
# Take an integer n and find the sum of its digits. (123=1+2+3)
n=input("Enter a number:")
count=0
for i in n:
  count=count+int(i)
print(count)
```

```python
# Take an integer n and print numbers from n to 1.
n = int(input("Enter a number: "))
for i in range(n, 0, -1):
    print(i)
```
