```python
# Largest Number in List Given:
nums = [4, 7, 1, 9, 3]
largest=nums[0]
for i in nums:
  if i>largest:
    largest=i
print(largest)
```

```python
# Take an integer n and count how many digits are greater than 5
n = int(input("Enter number: "))
count = 0

while n > 0:
    digit = n % 10
    if digit > 5:
        count += 1
    n //= 10

print(count)
```

```python
# nums = [1, 2, 3, 4] Create a new list in reverse order.
nums = [1, 2, 3, 4]
rev_list = nums[::-1]
print(rev_list) # Output: [4, 3, 2, 1]
```

```python
# nums = [3, 6, 9, 2, 8, 1] Find the sum of all even numbers.
nums = [3, 6, 9, 2, 8, 1]
sum=0
for i in nums:
  if(i%2==0):
    sum+=i
print(sum)
```
