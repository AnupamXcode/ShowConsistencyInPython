```python
# Take an integer n and count how many digits it has.
n = int(input("Enter a number: "))
count = 0
for i in str(abs(n)):
    count += 1
print(count)
```

```python
# Take an integer n and find the largest digit.
n = int(input("Enter number: "))
largest = 0
while n > 0:
    digit = n % 10
    if digit > largest:
        largest = digit
    n //= 10
print("Largest digit:", largest)
```

```python
'''
****
***
**
*
'''
n=int(input("Enter a number:"))
for i in range(n,0,-1):
  for j in range(i):
    print("*",end=" ")
  print(" ")
```
