```python
# Q1 — Smallest Number in List
nums = [7, 2, 10, 4, 1]
smallest = nums [0]
for i in nums:
  if i < smallest:
    smallest = i
print(smallest)
```

```python
# Q2 — Second Largest in List
nums = [4, 7, 2, 9, 5]
first = second = float('-inf')
for x in nums:
  if x > first:
    second = first
    first = x
  elif first > x > second:
    second = x

result = second if second != float('-inf') else "No second largest"
print(result) # Output: 7
```

```python
# Take an integer n and count how many digits are even.
n = int(input("Enter a number:"))
count = 0
temp = n
if temp == 0:
    count = 1
else:
    while temp > 0:
        digit = temp % 10
        if digit % 2 == 0:
            count += 1
        temp = temp // 10
print(count)
```

```python
# Take an integer n and reverse it.
n = int(input("Enter a number:"))
rev = 0
while n > 0:
  digit = n % 10
  rev =  rev *10 + digit
  n = n // 10
print(rev)
```
