```python
# Find the smallest number without using min().
nums = [7, 2, 10, 4, 1]
smallest = nums[0]
for i in nums:
  if i < smallest:
    smallest = i
print(smallest)
```

```python
# Find the Largest element without using Max()
nums = [1, 2, 5, 7, 10, 23, 17]
largest = nums[0]
for i in nums:
  if i > largest:
    largest=i
print(largest)
```

```python
# Count Odd Numbers in List
nums = [3, 6, 9, 2, 8, 1, 7]
count = 0
for i in nums:
  if i % 2 !=0:
    count+=1
print(count)
```

```python
# Check Digit Exists - Take an integer n and check whether digit 5 exists in the number.
n=int(input("Enter a number to Check:"))
found = False
while n > 0:
  digit = n % 10
  if digit == 5:
    found = True
    break
  n = n//10
if found:
  print("Yes")
else :
  print("No")
```
