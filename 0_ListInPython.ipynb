```python
# ==============================
# PYTHON LIST FUNCTIONS DEMO
# ==============================

# Initial List
numbers = [5, 2, 9, 1, 5, 6]
print("Initial List:", numbers)
print("-" * 50)

# 1️⃣ append() → Adds a single element at the end
numbers.append(10)
print("After append(10):", numbers)
print("append() adds one element to the end.\n")

# 2️⃣ insert(index, value) → Inserts at specific position
numbers.insert(2, 15)
print("After insert(2, 15):", numbers)
print("insert() adds element at a specific index.\n")

# 3️⃣ extend(iterable) → Adds multiple elements at end
numbers.extend([20, 25, 30])
print("After extend([20, 25, 30]):", numbers)
print("extend() adds multiple elements.\n")

# 4️⃣ remove(value) → Removes first occurrence of value
numbers.remove(5)
print("After remove(5):", numbers)
print("remove() deletes first matching value.\n")

# 5️⃣ pop(index) → Removes and returns element
popped_item = numbers.pop(3)
print("After pop(3):", numbers)
print("Popped Item:", popped_item)
print("pop() removes element by index and returns it.\n")

# 6️⃣ index(value) → Returns index of first occurrence
index_of_6 = numbers.index(6)
print("Index of 6:", index_of_6)
print("index() finds position of value.\n")

# 7️⃣ count(value) → Counts occurrences
count_5 = numbers.count(5)
print("Count of 5:", count_5)
print("count() returns how many times value appears.\n")

# 8️⃣ sort() → Sorts list in ascending order
numbers.sort()
print("After sort():", numbers)
print("sort() arranges list in ascending order.\n")

# 9️⃣ reverse() → Reverses the list
numbers.reverse()
print("After reverse():", numbers)
print("reverse() flips the list order.\n")

# 🔟 copy() → Creates a shallow copy
new_numbers = numbers.copy()
print("Copy created:", new_numbers)
print("copy() creates a separate list.\n")

# 1️⃣1️⃣ clear() → Removes all elements
numbers.clear()
print("After clear():", numbers)
print("clear() empties the list.\n")

print("=" * 50)
print("FINAL STATE")
print("Original list:", numbers)
print("Copied list:", new_numbers)
```

```python
# Print Positive and negative element of list Sepertely
main = [25, 65 ,56, -65, -68, 98, -67]
posi_list = []
neg_list = []
for i in range(len(main)):
  if main[i] > 0:
    posi_list.append(i)
  else:
    neg_list.append(i)
print(f"The Indexes where Positive numbers are there is : {posi_list}")
print(f"The Indexes where Negative numbers are there is : {neg_list}")
```

```python
# Mean of List elements
l = [23, 25, 56, 68, 10]
sum = 0
for i in l:
  sum += i
ans = sum // len(l)
print(float(ans))
```

```python
# Largest element in the List
# Pehle element ko largest consider krloo and compare krte rahooo
l = [23, 25, 56, 68, 10]
largest = l[0]
for i in range(len(l)):
  if l[i]>largest:
    largest = l[i]
    index = i
  else:
    largest = largest
print(f"The largest element is :{largest} At index :{index}")
```

```python
# Second largest element in List
# First value ko Largest maan lo , or usko hii second largest maan lo...fir ek ek krke change kroo or dekhoo koi sa agar bada hai to usko largest krdo and fir iterate kro

l = [25, 12, 56, 45, 55]
largest = second = l[0]
for i in l:
  if i >largest:
    second = largest
    largest = i
  elif i > second:
    second = i
print(largest,second)
```

```python
# Smallest number in a list with index
l = [22, 4, 54, 19, 43]
smallest = l[0]
for i in range(len(l)):
  if l[i] < smallest:
    smallest = l[i]
    index = i
  else :
    smallest = smallest
print(smallest, "at index",index)
```

```python
l = [22, 43, 34, 56, 23]
smallest = float('inf')
second_smallest = float('inf')

for i in l:
    if i < smallest:
        second_smallest = smallest
        smallest = i
    elif i < second_smallest and i != smallest:
        second_smallest = i

print("Smallest:", smallest)
print("Second Smallest:", second_smallest)
```
