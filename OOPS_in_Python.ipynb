```python
# Classes in OOPS
print("Class is like a blueprint or template for creating objects...")
print("Classes have two things: i) Attribute (Variables inside class) ii)Methods(Functions inside class)")
print(end="\n")
class Factory:
  a = 45 # Attribute haii ye
  def new():
    print("Hello How are youuu ? i am getting an Internship")
  print("Are mujhe bhi initialise kr diaa woww...")

# For accessing a inside the Factory we need to call factory as
print(Factory().a)
```

```python
class Sum:
  a = 45
  b = 54
  print(a+b)
  def sum (a,b):
    print(f"The sum of {a} and {b} is {a+b}")

a = Sum # ye class ka object bana diya hamne
print(a.sum(5,6))  # accessing function using object
```
