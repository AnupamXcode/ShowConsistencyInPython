```python
# Take an integer n and print numbers from 1 to n.
n=int(input("Enter a Number :"))
for i in range(1,n+1):
  print(i)
```

```python
# Take an integer n and find the sum of all odd numbers from 1 to n.
n=int(input("Enter a number:"))
total=0
for i in range(1,n+1):
  if (i%2!=0):
    total+=i
print(total)
```

```python
# Take an integer n and count how many even numbers exist from 1 to n.
n=int(input("Enter a number:"))
count=0
for i in range(1,n+1):
  if(i%2==0):
    count+=1
print(count)
```
